## <font color='lightblue'>2. Summary statistics</font>

**Action:** Estimate the summary statistics of some of the key variables

```
alcohol_mean = df_wine['alcohol'].mean()
alcohol_mode = df_wine['alcohol'].mode()
alcohol_median = df_wine['alcohol'].median()
alcohol_std = df_wine['alcohol'].std()

print("The mean alcohol content is: ", alcohol_mean)
print("The most common alcohol content is: ", alcohol_mode)
print("The median alcohol content is: ", alcohol_median)
print("The standard deviation of alcohol contents is: ", alcohol_std)
```

```
hue_mean = df_wine['hue'].mean()
hue_mode = df_wine['hue'].mode()
hue_median = df_wine['hue'].median()
hue_std = df_wine['hue'].std()

print("The mean hue is: ", hue_mean)
print("The most common hue is: ", hue_mode)
print("The median hue is: ", hue_median)
print("The standard deviation of hues is: ", hue_std)
```

```
sugar_mean = df_wine['total_sugar'].mean()
sugar_mode = df_wine['total_sugar'].mode()
sugar_median = df_wine['total_sugar'].median()
sugar_std = df_wine['total_sugar'].std()

print("The mean sugar content is: ", sugar_mean)
print("The most common sugar content is: ", sugar_mode)
print("The median sugar content is: ", sugar_median)
print("The standard deviation of sugar contents is: ", sugar_std)
```

## <font color='lightblue'>3. Visualize the data</font>

**Action:** Visualize the distribution of values for some key variables

```
sns.displot(df_wine, x='alcohol', binwidth = 0.45)
plt.axvline(alcohol_mean, color="red")
plt.axvline(alcohol_mode[0], color="blue")
plt.axvline(alcohol_median, color="orange")
```

```
sns.displot(df_wine, x='total_sugar')
plt.axvline(sugar_mean, color="red")
plt.axvline(sugar_mode[0], color="blue")
plt.axvline(sugar_median, color="orange")
```

```
sns.displot(df_wine, x='flavanoids', kind='kde')
```

```
df_wine_pairs = df_wine[["alcohol", "hue", "total_sugar"]]

sns.pairplot(df_wine_pairs)
```

```
sns.scatterplot(data=df_wine, x='flavanoids', y='hue').set(xlabel='Flavanoids', ylabel='Hue')
```
